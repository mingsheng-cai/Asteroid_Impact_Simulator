```python
import pandas as pd
import sys
import deepimpact
sys.path.append('./resources')
df = pd.read_csv("./resources/impact_parameter_list_100.csv")
mf = deepimpact.mapping.MappingFunctions()
earth = deepimpact.Planet()
columns = ['lat0', 'lon0', 'lat1', 'lon1', 'radius1', 'radius2', 'radius3', 'radius4']
data = pd.DataFrame(columns=columns)
for index, row in df.head(3).iterrows():    
    result = earth.solve_atmospheric_entry(
        radius=row['radius'], angle=row['angle'], strength=row['strength'],
        density=row['density'], velocity=row['velocity']
    )
    # Calculate the kinetic energy lost per unit altitude and add it
    result = earth.calculate_energy(result)
    # Determine the outcomes of the impact event
    outcome = earth.analyse_outcome(result)
    # Pressure levels
    pressures = [1e3, 4e3, 30e3, 50e3]

    blast_lat, blast_lon, damage_rad = deepimpact.damage_zones(
        outcome, lat=row['entry latitude'], lon=row['entry longitude'], bearing=row['bearing'], pressures=pressures
    )
    data1= mf.create_map_df(row['entry latitude'], row['entry longitude'], blast_lat, blast_lon, damage_rad)
    data = pd.concat([data, data1], ignore_index=True)
# data
mf.plot_damage_areas(data)
fmap = mf.get_fmap()
fmap
```
